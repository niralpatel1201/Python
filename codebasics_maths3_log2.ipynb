```python
df = pd.read_csv('income_log.csv', index_col = None, names =["income", "count"], skiprows=1)
df.head()
```

1. sns.set(rc={'figure.figsize':(11.7, 8.27)})
sns.set(): This sets the aesthetic style for Seaborn plots.
rc: A dictionary of parameters to customize the appearance of the plot.
figure.figsize: Specifies the size of the figure in inches:
(11.7, 8.27) means the plot will be 11.7 inches wide and 8.27 inches tall.
2. g = sns.barplot(x='income', y='count', data=df)
sns.barplot(): Creates a bar plot to show the relationship between a categorical variable (x-axis) and a numeric variable (y-axis).
x='income': The column income in the DataFrame df is used for the x-axis. It likely contains categories (e.g., income ranges like "0-10k").
y='count': The column count in df is used for the y-axis. It likely contains numerical values (e.g., number of people in each income range).
data=df: Specifies the DataFrame (df) containing the data for the plot.
The plot is stored in the variable g, which is a Seaborn AxesSubplot object. This allows further customization of the plot.
3. g.set_xticklabels()
This customizes the labels on the x-axis (income categories in this case).
g.get_xticklabels(): Retrieves the current tick labels on the x-axis.
rotation=45: Rotates the x-axis labels by 45 degrees for better readability.
horizontalalignment='right': Aligns the rotated labels to the right for a cleaner appearance.

```python
sn.set(rc={'figure.figsize':(11.7, 8.27)})
g= sn.barplot(x='income', y ='count', data=df)
g.set_xticklabels(g.get_xticklabels(), rotation=45, horizontalalignment='right');
```

```python
sn.set(rc={'figure.figsize':(11.7,8.27)})
g = sn.barplot(x='income',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(), 
                          rotation=45, 
                          horizontalalignment='right');
g.set(xscale="log");
```
